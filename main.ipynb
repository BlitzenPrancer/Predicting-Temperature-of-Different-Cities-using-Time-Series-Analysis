```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
 global_temp = pd.read_csv(r'C:\Users\user\code\datasets\Predicting the temperatures\GlobalTemperatures.csv')
```

```python
global_temp.head()
```

```python
global_temp['dt'][0].split('-')[0]
```

```python
def fetch_year(date):
    return date.split('-')[0]
```

```python
global_temp['years'] = global_temp['dt'].apply(fetch_year)
```

```python
global_temp.head()
```

```python
# on basis of years grouping the data 
# and converting it into a dataframe using reset_index()
data = global_temp.groupby('years').agg({'LandAverageTemperature': 'mean', 'LandAverageTemperatureUncertainty': 'mean'}).reset_index()
```

```python
data
```

```python
# adding derived attributes uncertainity_top and uncertainity_bottom
data['uncertainity_top'] = data['LandAverageTemperature'] + data['LandAverageTemperatureUncertainty']
data['uncertainity_bottom'] = data['LandAverageTemperature'] - data['LandAverageTemperatureUncertainty']
```

```python
data.head()
```

```python
# plotting to check if global warming exists of not
fig = px.line(data, x = 'years', y = ['LandAverageTemperature', 'uncertainity_top', 'uncertainity_bottom'], title = 'Average land temperature')
fig.show()
```

From the above we can observe the avereage temperature on Earth's surface has the highest over the last decades
